```python
import sys
import collections as col

graph={}
def make_graph(ele1, ele2): #인접리스트 사용
    if ele1 in graph:
        if ele2 in graph:
            graph[ele1].add(ele2)
            graph[ele2].add(ele1)
        else:
            graph[ele1].add(ele2)
            graph[ele2]={ele1}
    else:
        if ele2 in graph:
            graph[ele1]={ele2}
            graph[ele2].add(ele1)
        else:
            graph[ele1]={ele2}
            graph[ele2]={ele1}


def dfs(graph, start_node, visited=set()):
    if start_node not in visited:
        visited.add(start_node)
        print(start_node, end=' ')
        unvisited_node=graph[start_node] - visited
        for i in unvisited_node:
            dfs(graph, i, visited)


def bfs(graph, start_node, visited=set()):
    queue=col.deque([start_node])
    visited.add(start_node)
    while len(queue)!=0:
        poped_node=queue.popleft()
        visited.add(poped_node)
        print(poped_node, end=' ')
        unvisited_node=graph[poped_node] - visited
        for i in unvisited_node:
            queue.append(i)
            visited.add(i)


node_cnt, link_cnt, start_node= map(int, sys.stdin.readline().split())
for i in range(link_cnt):
    ele1, ele2 = map(int, sys.stdin.readline().split())
    make_graph(ele1, ele2)

dfs(graph, start_node)
print()
bfs(graph, start_node)
```

```python
import sys
import collections as col

graph={}
def make_graph(ele1, ele2): #인접리스트 사용; { ele1: [ele2] }
    if ele1 in graph: # ele1가 이미 입력된 경우
        if ele2 in graph: # ele2도 이미 입력된 경우
            graph[ele1].append(ele2) # ele1의 연결리스트에 ele2 추가
            graph[ele2].append(ele1) # 위와 같이 ele2에도 하기

        else: # ele2가 처음으로 입력된 경우
            graph[ele1].append(ele2)
            graph[ele2]=[ele1] # ele2의 연결리스트 만들기

    else: # ele1가 처음으로 입력된 경우
        if ele2 in graph:
            graph[ele1]=[ele2]
            graph[ele2].append(ele1)

        else:
            graph[ele1]=[ele2]
            graph[ele2]=[ele1]

# dfs
# 이미 방문한 노드는 visited에 추가
# start_node의 인접리스트 중 방문하지 않은 노드를 start_node로 하여 dfs 순환
# 인접리스트는 이미 정렬되어 있음; 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문
def dfs(graph, start_node, visited=[]):
    if start_node not in visited:
        visited.append(start_node)
        print(start_node, end=' ')
        for i in graph[start_node]:
            if i not in visited:
                dfs(graph, i, visited)

# bfs
# 시작노드를 큐와 visited에 추가
# 큐의 첫번째 노드를 빼서 그 노드의 인접리스트들 중 방문한 적이 없는 노드를 큐에 추가
# 위의 과정을 큐가 빌 때까지 반복
# 인접리스트는 이미 정렬되어 있음
def bfs(graph, start_node, visited=[]):
    queue=col.deque([start_node])
    visited.append(start_node)
    while len(queue)>0:
        poped_node=queue.popleft()
        print(poped_node, end=' ')
        for i in graph[poped_node]:
            if i not in visited:
                queue.append(i)
                visited.append(i)



node_cnt, link_cnt, start_node= map(int, sys.stdin.readline().split())
for i in range(link_cnt):
    ele1, ele2 = map(int, sys.stdin.readline().split())
    make_graph(ele1, ele2)

for i in graph: # 그래프의 인접리스트를 정렬
    graph[i].sort()

dfs(graph, start_node)
print()
bfs(graph, start_node)
```

```python
import sys
import collections as col


node_cnt, link_cnt, start_node= map(int, sys.stdin.readline().split())

graph=[[0]*node_cnt for i in range(node_cnt)]
def make_graph(ele1, ele2): #인접행렬 사용;
    graph[ele1-1][ele2-1] = 1
    graph[ele2-1][ele1-1] = 1


# dfs
# 이미 방문한 노드는 visited에 추가
# start_node의 인접리스트 중 방문하지 않은 노드를 start_node로 하여 dfs 순환
# 인접리스트는 이미 정렬되어 있음; 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문
def dfs(start_node, visited=[]):
    if start_node not in visited:
        visited.append(start_node)
        print(start_node, end=' ')
        for i in range(node_cnt):
            if graph[start_node-1][i]!=0 and i+1 not in visited:
                dfs(i+1,visited)

# bfs
# 시작노드를 큐와 visited에 추가
# 큐의 첫번째 노드를 빼서 그 노드의 인접리스트들 중 방문한 적이 없는 노드를 큐에 추가
# 위의 과정을 큐가 빌 때까지 반복
# 인접리스트는 이미 정렬되어 있음
def bfs(start_node, visited=[]):
    queue=col.deque([start_node])
    visited.append(start_node)
    while len(queue)>0:
        poped_node=queue.popleft()
        print(poped_node, end=' ')
        for i in range(node_cnt):
            if graph[poped_node -1][i] and i+1 not in visited:
                queue.append(i+1)
                visited.append(i+1)



for i in range(link_cnt):
    ele1, ele2 = map(int, sys.stdin.readline().split())
    make_graph(ele1, ele2)

dfs(start_node)
print()
bfs(start_node)
```

```python
import sys
import collections as col

map1=[]
map_len=int(sys.stdin.readline())
for i in range(map_len):
    row1 = list(map(int, sys.stdin.readline().strip()))
    map1.append(row1)

visited=[] # 방문했던 위치를 저장하는 리스트

def bfs(map1, i, j): # bfs 탐색
    visited.append((i,j))
    queue=col.deque([(i,j)])
    cnt = 1 # 단지내 집의 개수 
    while len(queue) !=0:
        unit=queue.popleft() # 유닛(집의 위치)을 하나씩 꺼내기
        i=unit[0]
        j=unit[1]
        unit_around=[] # 현재 유닛의 상 하 좌 우 유닛들 중 값이 1인 유닛을 저장하는 리스트
        if i!=0: # 상
            if map1[i-1][j]==1:
                unit_around.append((i-1, j))
        if j!=0: # 좌
            if map1[i][j-1] == 1:
                unit_around.append((i, j-1))
        if i!=map_len-1: # 하
            if map1[i+1][j] == 1:
                unit_around.append((i+1, j))
        if j!=map_len-1: # 우
            if map1[i][j+1] == 1:
                unit_around.append((i, j+1))

        for u in unit_around:
            if u not in visited: # 방문한 적이 없다면
                visited.append(u)
                queue.append(u)
                cnt+=1 # 단지내 집의 개수 +1
    return cnt # 단지내 집의 개수  반환


def search_apartment(map1):
    apartment_len=[]# 각 단지내 집의 수를 저장하는 리스트
    for i in range(map_len):
        for j in range(map_len): # map의 모든 요소를 방문
            if map1[i][j] ==1 and (i,j) not in visited: # 만약 현재 요소가 1이고 방문한 적이 없다면
                apartment_len.append(bfs(map1,i,j)) # bfs 탐색 시작-> 현재요소와 연결되어 있고 1인 요소들을 탐색
    apartment_len.sort() # 각 단지내 집의 수들을 정렬
    print(len(apartment_len)) #총 단지 수
    for i in apartment_len:
        print(i)

search_apartment(map1)



```
