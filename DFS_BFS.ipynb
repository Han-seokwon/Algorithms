```python
import sys
import collections as col

graph={}
def make_graph(ele1, ele2): #인접리스트 사용
    if ele1 in graph:
        if ele2 in graph:
            graph[ele1].add(ele2)
            graph[ele2].add(ele1)
        else:
            graph[ele1].add(ele2)
            graph[ele2]={ele1}
    else:
        if ele2 in graph:
            graph[ele1]={ele2}
            graph[ele2].add(ele1)
        else:
            graph[ele1]={ele2}
            graph[ele2]={ele1}


def dfs(graph, start_node, visited=set()):
    if start_node not in visited:
        visited.add(start_node)
        print(start_node, end=' ')
        unvisited_node=graph[start_node] - visited
        for i in unvisited_node:
            dfs(graph, i, visited)


def bfs(graph, start_node, visited=set()):
    queue=col.deque([start_node])
    visited.add(start_node)
    while len(queue)!=0:
        poped_node=queue.popleft()
        visited.add(poped_node)
        print(poped_node, end=' ')
        unvisited_node=graph[poped_node] - visited
        for i in unvisited_node:
            queue.append(i)
            visited.add(i)


node_cnt, link_cnt, start_node= map(int, sys.stdin.readline().split())
for i in range(link_cnt):
    ele1, ele2 = map(int, sys.stdin.readline().split())
    make_graph(ele1, ele2)

dfs(graph, start_node)
print()
bfs(graph, start_node)
```

# 2667번 단지번호붙이기

- 맞은 코드

```python
import sys
import collections as col

map1=[]
map_len=int(sys.stdin.readline())
for i in range(map_len):
    row1 = list(map(int, sys.stdin.readline().strip()))
    map1.append(row1)

visited=[] # 방문했던 위치를 저장하는 리스트

def bfs(map1, i, j): # bfs 탐색
    visited.append((i,j))
    queue=col.deque([(i,j)])
    cnt = 1 # 단지내 집의 개수 
    while len(queue) !=0:
        unit=queue.popleft() # 유닛(집의 위치)을 하나씩 꺼내기
        i=unit[0]
        j=unit[1]
        unit_around=[] # 현재 유닛의 상 하 좌 우 유닛들 중 값이 1인 유닛을 저장하는 리스트
        if i!=0: # 상
            if map1[i-1][j]==1:
                unit_around.append((i-1, j))
        if j!=0: # 좌
            if map1[i][j-1] == 1:
                unit_around.append((i, j-1))
        if i!=map_len-1: # 하
            if map1[i+1][j] == 1:
                unit_around.append((i+1, j))
        if j!=map_len-1: # 우
            if map1[i][j+1] == 1:
                unit_around.append((i, j+1))

        for u in unit_around:
            if u not in visited: # 방문한 적이 없다면
                visited.append(u)
                queue.append(u)
                cnt+=1 # 단지내 집의 개수 +1
    return cnt # 단지내 집의 개수  반환


def search_apartment(map1):
    apartment_len=[]# 각 단지내 집의 수를 저장하는 리스트
    for i in range(map_len):
        for j in range(map_len): # map의 모든 요소를 방문
            if map1[i][j] ==1 and (i,j) not in visited: # 만약 현재 요소가 1이고 방문한 적이 없다면
                apartment_len.append(bfs(map1,i,j)) # bfs 탐색 시작-> 현재요소와 연결되어 있고 1인 요소들을 탐색
    apartment_len.sort() # 각 단지내 집의 수들을 정렬
    print(len(apartment_len)) #총 단지 수
    for i in apartment_len:
        print(i)

search_apartment(map1)



```

```python
import sys
import collections as col

# 입력
map1=[]
riped_tomato_loc=[]
width, height=map(int, sys.stdin.readline().split())
for i in range(height):
    row1 = list(map(int, sys.stdin.readline().split()))
    for j in range(width):
        if row1[j]==1:
            riped_tomato_loc.append((i,j))
    map1.append(row1)

def search_map(map1):
    for i in range(height):
        for j in range(width):
            if map1[i][j] == 0:
                return -1
    return 0

def tomato_riped_day(map1): # bfs
    visited=[]
    queue = col.deque()
    for t in riped_tomato_loc:
        visited.append(t)
        queue.append(t)

    while len(queue) != 0:
        unit = queue.popleft()
        i = unit[0] # y 축
        j = unit[1] # x 축

        day_cnt=map1[i][j] # 걸린 일

        if i > 0:  # 상
            if map1[i - 1][j] != -1 and (i-1,j) not in visited: # 위에 있는 점이 -1이 아니고 방문한 적이 없다면
                visited.append((i - 1, j))
                queue.append((i - 1, j))
                map1[i-1][j]= day_cnt + 1 # 현재 좌표의 카운트에서 +1을 대입

        if i < height - 1:  # 하
            if map1[i + 1][j] != -1 and (i+1,j) not in visited:
                visited.append((i + 1, j))
                queue.append((i + 1, j))
                map1[i + 1][j] = day_cnt + 1

        if j > 0:  # 좌
            if map1[i][j - 1] != -1 and (i,j-1) not in visited:
                visited.append((i, j-1))
                queue.append((i , j-1))
                map1[i][j-1] = day_cnt + 1

        if j < width - 1:  # 우
            if map1[i][j + 1] != -1 and (i,j+1) not in visited:
                visited.append((i, j+1))
                queue.append((i, j+1))
                map1[i][j+1] = day_cnt + 1

    if search_map(map1) == -1: # 0이 하나 라도 있는 경우
        print(-1)
    else:
        print(day_cnt -1) # day_cnt 가 1부터 시작해서 1을 뺴줌

tomato_riped_day(map1)

```

# 7569번 토마토

- 맞은 코드

```python
import sys
import collections as col

# 입력
map1=[]
riped_tomato_loc=[]
unriped_tomato_cnt=0 # 익지 않은 토마토 개수

width, height, layer=map(int, sys.stdin.readline().split())

for z in range(layer):
    floor = []
    for y in range(height):
        row1 = list(map(int, sys.stdin.readline().split()))
        for x in range(width):
            if row1[x]==1:
                riped_tomato_loc.append((z,y,x))
            elif row1[x]==0:
                unriped_tomato_cnt+=1
        floor.append(row1)
    map1.append(floor)


def tomato_riped_day(map1): # bfs

    global unriped_tomato_cnt

    queue = col.deque()
    for t in riped_tomato_loc:
        queue.append(t)
    max_day_cnt = 0
    while len(queue) != 0:
        unit = queue.popleft()

        v = unit[0] # z 축
        i = unit[1] # y 축
        j = unit[2] # x 축

        day_cnt=map1[v][i][j] # 걸린 일
        if day_cnt > max_day_cnt:
            max_day_cnt = day_cnt

        if v > 0: # 위
            if map1[v-1][i][j] == 0:
                queue.append((v-1,i, j))
                map1[v-1][i][j] = day_cnt + 1
                unriped_tomato_cnt -= 1

        if v < layer -1: #아래
            if map1[v+1][i][j] == 0:
                queue.append((v+1,i, j))
                map1[v+1][i][j] = day_cnt + 1
                unriped_tomato_cnt -= 1

        if i > 0:  # 앞
            if map1[v][i - 1][j] ==0: # 위에 있는 점이 0이고 방문한 적이 없다면
                queue.append((v,i - 1, j))
                map1[v][i-1][j]= day_cnt + 1 # 현재 좌표의 카운트에서 +1을 대입
                unriped_tomato_cnt -= 1 # 안 익은 토마토 개수 -1

        if i < height - 1:  # 뒤
            if map1[v][i + 1][j] ==0:
                queue.append((v,i + 1, j))
                map1[v][i + 1][j] = day_cnt + 1
                unriped_tomato_cnt -= 1


        if j > 0:  # 좌
            if map1[v][i][j - 1] ==0:
                queue.append((v,i , j-1))
                map1[v][i][j-1] = day_cnt + 1
                unriped_tomato_cnt -= 1

        if j < width - 1:  # 우
            if map1[v][i][j + 1] ==0:
                queue.append((v,i, j+1))
                map1[v][i][j+1] = day_cnt + 1
                unriped_tomato_cnt -= 1

    if unriped_tomato_cnt > 0: # 0이 하나 라도 있는 경우
        print(-1)
    else:
        print(max_day_cnt -1) # day_cnt 가 1부터 시작해서 1을 뺴줌

tomato_riped_day(map1)

```

```python
import sys
import collections as col


# 입력
start , end = map(int, sys.stdin.readline().split())

def shortest_time():
    if start < end:
        point = start
        visited = [point] # 이미 방문한 지점을 저장하는 리스트
        queue=col.deque([point])
        second=1
        # 매 초마다 가능한 조합은 3의 제곱씩 늘어남 3, 9, 27 ...
        cnt = 1  # 매 초마다 가능한 조합의 개수; 1부터 시작
        # 만약 해당 포인트가 이미 방문한 지점이라면 다음 초에서 가능한 조합이 하나 줄어야함
        visited_cnt=0 # 해당 포인트가 방문한 지점일 때 다음 초에서 빼 줄 cnt를 기록함
        while point != end:
            point=queue.popleft()

            if point+1 not in visited:
                queue.append(point + 1)
                visited.append(point + 1)
            else: # 이미 방문한 포인트면 이 후로 계산할 필요가 없음
                visited_cnt +=1

            if point - 1 not in visited:
                queue.append(point - 1)
                visited.append(point - 1)
            else:
                visited_cnt += 1

            if point * 2 not in visited:
                queue.append(point * 2)
                visited.append(point * 2)
            else:
                visited_cnt += 1

            if cnt == 0: # 해당 시간(초)에 일어날 수 있는 모든 조합을 다 거친 경우
                # 매 초의 첫번째 조합마다 second가 늘어남
                cnt = 3**second # 다음 시간대에 일어날 수 있는 모든 조합의 개수; 3제곱씩 늘어남
                cnt -= visited_cnt # 가능한 조합에서 이미 방문해 연산하지 않아도 되는 조합을 빼 줌
                visited_cnt = 0
                second += 1

            cnt-=1 # 가능한 조합의 개수 -1

        return second - 1 # 1초 부터 시작 했으므로 1을 빼줌

    elif start == end:
        return 0

    else:
        return start - end


print(shortest_time())
```
