# 우선순위 큐
- 우선순위 큐의 가장 효율적인 방법은 트리 구조를 사용하는 Heap이다.

- 성공한 코드

```python
import sys
# 최대힙을 클래스로 구현
class maxHeap:
    # 부모 위치 = 자식 위치//2
    # 왼쪽 자식위치 = 부모 위치*2
    # 오른쪽 자식위치 = 부모 위치*2 +1
    def __init__(self):
        self.heap=[0] # 힙의 첫번째 요소는 0; 비어 있는 항목; 힙의 인덱스 0부터 시작하면 부모와 자식의 위치가 겹쳐짐

    def isEmpty(self):
        return len(self.heap)==1 # 힙의 첫번 째 칸은 비어 있음

    def push(self, ele): # 추가 함수
        if self.isEmpty(): # 비어있는 경우는 바로 원소 추가
            self.heap.append(ele)
           
        else: # 힙이 안 비어 있는 경우
            self.heap.append(ele) # 우선 마지막에 원소 추가
            ele_index = len(self.heap) - 1 # 추가된 원소의 위치
            parent=self.heap[ele_index // 2] # 추가된 원소의 부모
            
            while ele_index!=1 and ele>parent: # 추가된 원소가 루트에 도달하지 않았고 부모 원소보다 큰 경우(힙 트리에서 올라가야 함)
                self.heap[ele_index]=parent # 부모를 추가된 원소의 위치로 끌어내림
                ele_index= ele_index // 2 # 추가된 원소의 위치를 부모의 위치로 옮김
                parent = self.heap[ele_index // 2] # 새로운 부모의 값 갱신
                
            self.heap[ele_index]=ele # 추가된 원소의 최종 위치에 원소 넣기

    def delete(self): # 삭제 함수
        if self.isEmpty(): # 힙이 비어 있는 경우
            return 0

        else:# 힙이 안 비어 있는 경우

            # 루트노드를 제거하고 가장 뒤에 있는 원소를 빼서 루트노드에 추가
            # 이후 자식노드와 비교하며 가장 큰 자식보다 더 작으면 밑으로 내려옴, 크면 거기서 멈춤

            max_ele=self.heap[1] # 현재 힙의 최댓값; 후에 출력
            node=self.heap.pop() # 힙의 가장 뒤의 값인 기준 노드(힙의 원소가 1개인 경우 그대로 삭제됨)
            node_index=1 # 기준 노드가 루트로 올라갔다고 가정하고 그떄의 인덱스

            while len(self.heap)-1>=node_index: # 기준 노드의 위치가 맨 뒤로 갈 때까지 반복
                
                if len(self.heap)-1>=node_index*2+1: # 기준 노드의 양쪽 자식이 모두 있는 경우
                    
                    if self.heap[node_index*2]> self.heap[node_index*2+1]: # 양쪽 자식 중 왼쪽 자식이 더 큰 경우
                        child = self.heap[node_index*2] # 가장 큰 자식= 왼쪽노드
                        child_index=node_index*2 # 자식 위치
                        
                    else: # 양쪽 자식 중 오른쪽 자식이 더 큰 경우
                        child = self.heap[node_index * 2+1] #가장 큰 자식= 오른쪽노드
                        child_index = node_index * 2 +1
                    
                    # 자식과 기준노드를 비교하는 과정
                    if node>=child: # 기준노드가 가장 큰 자식보다 크거나 같은 경우
                        break # 멈춤
                        
                    else: # 기준노드가 가장 큰 자식보다 작은 경우
                        self.heap[node_index] = child # 기준노드의 자리로 자식이 올라감
                        node_index=child_index # 기준노드의 위치는 자식의 자리로 내려감
                        
                elif len(self.heap)-1==node_index*2: # 기준 노드의 왼쪽 자식만 있는 경우(완전이진트리여서 오론쪽 자식만 있는 경우는 없음)
                    child = self.heap[node_index*2]
                    child_index = node_index * 2
                    # 자식과 기준노드를 비교하는 과정
                    if node>=child:
                        break
                    else:
                        self.heap[node_index] = child
                        node_index = child_index
                else:
                    break

            if not self.isEmpty(): # 만약 노드가 빈 경우(원래 원소가 1개 있었음)는 그대로 사라짐
                self.heap[node_index] = node # 최종적으로 결정된 기준 노드의 위치에 대입

            return max_ele # 위에서 저장한 가장 큰 값 반환

heap=maxHeap()
for i in range(int(sys.stdin.readline())): # 입력 출력
    order=int(sys.stdin.readline())
    if order == 0:
        print(heap.delete())
    else:
        heap.push(order)
```

```python
import sys
# 최소힙을 클래스로 구현
class minHeap:
    # 부모 위치 = 자식 위치//2
    # 왼쪽 자식위치 = 부모 위치*2
    # 오른쪽 자식위치 = 부모 위치*2 +1
    def __init__(self):
        self.heap=[0] # 힙의 첫번째 요소는 0; 비어 있는 항목; 힙의 인덱스 0부터 시작하면 부모와 자식의 위치가 겹쳐짐

    def isEmpty(self):
        return len(self.heap)==1 # 힙의 첫번 째 칸은 비어 있음

    def push(self, ele): # 추가 함수
        if self.isEmpty(): # 비어있는 경우는 바로 원소 추가
           self.heap.append(ele)
           
        else: # 힙이 안 비어 있는 경우
            self.heap.append(ele) # 우선 마지막에 원소 추가
            ele_index = len(self.heap) - 1 # 추가된 원소의 위치
            parent=self.heap[ele_index // 2] # 추가된 원소의 부모
            
            while ele_index!=1 and ele<parent: # 추가된 원소가 루트에 도달하지 않았고 부모 원소보다 작은 경우(힙 트리에서 올라가야 함)
                self.heap[ele_index]=parent # 부모를 추가된 원소의 위치로 끌어내림
                ele_index= ele_index // 2 # 추가된 원소의 위치를 부모의 위치로 옮김
                parent = self.heap[ele_index // 2] # 새로운 부모의 값 갱신
                
            self.heap[ele_index]=ele # 추가된 원소의 최종 위치에 원소 넣기

    def delete(self): # 삭제 함수
        if self.isEmpty(): # 힙이 비어 있는 경우
            return 0

        else:# 힙이 안 비어 있는 경우

            # 루트노드를 제거하고 가장 뒤에 있는 원소를 빼서 루트노드에 추가
            # 이후 자식노드와 비교하며 가장 작은 자식보다 더 크면 밑으로 내려옴, 작으면 거기서 멈춤

            min_ele=self.heap[1] # 현재 힙의 최솟값; 후에 출력
            node=self.heap.pop() # 힙의 가장 뒤의 값인 기준 노드(힙의 원소가 1개인 경우 그대로 삭제됨)
            node_index=1 # 기준 노드가 루트로 올라갔다고 가정하고 그떄의 인덱스

            while len(self.heap)-1>=node_index: # 기준 노드의 위치가 맨 뒤로 갈 때까지 반복
                
                if len(self.heap)-1>=node_index*2+1: # 기준 노드의 양쪽 자식이 모두 있는 경우
                    
                    if self.heap[node_index*2]< self.heap[node_index*2+1]: # 양쪽 자식 중 왼쪽 자식이 더 작은 경우
                        child = self.heap[node_index*2] # 가장 작은 자식= 왼쪽노드
                        child_index=node_index*2 # 자식 위치
                        
                    else: # 양쪽 자식 중 오른쪽 자식이 더 큰 경우
                        child = self.heap[node_index * 2+1] #가장 작은 자식= 오른쪽노드
                        child_index = node_index * 2 +1
                    
                    # 자식과 기준노드를 비교하는 과정
                    if node<=child: # 기준노드가 가장 작은 자식보다 작거나 같은 경우
                        break # 멈춤
                        
                    else: # 기준노드가 가장 작은 자식보다 작은 경우
                        self.heap[node_index] = child # 기준노드의 자리로 자식이 올라감
                        node_index=child_index # 기준노드의 위치는 자식의 자리로 내려감
                        
                elif len(self.heap)-1==node_index*2: # 기준 노드의 왼쪽 자식만 있는 경우(완전이진트리여서 오론쪽 자식만 있는 경우는 없음)
                    child = self.heap[node_index*2]
                    child_index = node_index * 2

                    # 자식과 기준노드를 비교하는 과정
                    if node<=child:
                        break
                    else:
                        self.heap[node_index] = child
                        node_index = child_index
                else:
                    break

            if not self.isEmpty(): # 만약 노드가 빈 경우(원래 원소가 1개 있었음)는 그대로 사라짐
                self.heap[node_index] = node # 최종적으로 결정된 기준 노드의 위치에 대입

            return min_ele # 위에서 저장한 가장 작은 값 반환

heap=minHeap()
for i in range(int(sys.stdin.readline())): # 입력 출력
    order=int(sys.stdin.readline())
    if order == 0:
        print(heap.delete())
    else:
        heap.push(order)
```
